```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph1 = nx.DiGraph([(0,1),(1,2),(2,3),(3,0)])
print(graph1.nodes())
print(graph1.edges())
```

```python
# Grafo não direcionado
graph_not_directed = nx.Graph([(0,1),(1,2),(2,3),(3,0)])
nx.draw(graph_not_directed, with_labels=True)
```

```python
grapho2 = nx.Graph()
grapho2.add_nodes_from(range(11))
grapho2.add_edges_from([(0,1),(2,2),(1,2),(2,3),(3,3)])
arestas = grapho2.nodes()
vertices = grapho2.edges()
plt.title(f"Número de arestas {len(arestas)} e vertices {len(vertices)}")
nx.draw(grapho2, with_labels= True, alpha=0.5)
```

```python
graph = nx.Graph()
graph.add_nodes_from(range(6))
graph.add_edges_from([(0,1),(1,2),(2,3),(3,4), (4,4)])
graph.add_node(6)
graph.add_node(7)
options = {"edgecolors": "tab:gray", "alpha": 0.9}
nx.draw(graph, **options, labels={0:'a',1:"b",2:"c",3:"d",4:"e"})
```
