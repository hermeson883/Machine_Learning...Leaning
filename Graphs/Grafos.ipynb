```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Definição do grafo direcionado com arestas
G = nx.DiGraph([(0,1),(1,2),(2,3),(3,0)], names = "My firts Graph")
print(G.number_of_edges()) # Número de arestas
print(G.number_of_nodes()) # Número de vertices
```

```python
nx.draw(G)
```

```python
# Segundo grafo
G2 = nx.DiGraph([(0,1),(1,2),(2,3),(3,0),(0,3),(3,2),(2,1),(1,0)])
plt.title("Grafo bidirecional")
nx.draw(G2, with_labels=True)
plt.show()
```

```python
#Grafo com laço
G3 = nx.DiGraph([(0,1),(1,2),(2,3),(3,0),(1,1)])
plt.title("Grafo com laço")
nx.draw(G3, with_labels=True)
plt.show()
```

```python
# Criação do quarto grafo
G4 = nx.Graph([(0,1),(1,2),(2,3),(3,0),(1,1)])
plt.title("Vertice isolado")
#Adicionando um vértice
G4.add_node(5)
nx.draw(G4, with_labels=True)
```
