```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
import pandas as pd
from scipy import stats
```

```python
irisDataset = pd.read_csv("iris.csv")
irisDataset
```

```python
iris = datasets.load_iris()
stats.describe(iris.data)
```

```python
iris.data
```

```python
previsores = iris.data # Variáveis independentes
classes = iris.target # Variável depêndente
```

```python
X_treinamento, X_teste, y_train, y_test = train_test_split(previsores, classes, test_size=0.3, random_state = 42)
```

```python
#Criação de modelo
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_treinamento, y_train)
```

```python
previsoes = knn.predict(X_teste)
previsoes
```

```python
#accuracy
accuracy_score(y_test,previsoes)
```

```python
confusion_matrix(y_test, previsoes)
```
