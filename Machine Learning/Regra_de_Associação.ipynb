```python
import pandas as pd
from apyori import apriori
```

```python
dados = pd.read_csv("transacoes.txt", header = None)
dados
```

```python
#Transformação para o formato de lista
transacoes = []

for i in range(6):
    transacoes.append([ str(dados.values[i,j]) for j in range(3)])
transacoes
```

```python
'''
    Algoritmo apriori()

    transactions = datas for the algorithm

    min_support = É uma porcentagem que temos que colocar para falar qual a frequência que queremos encotrar os nossos items

    min_confidence = É uma correlação para ver qual a frequência de vezes que os itens são vistos juntos

    min_lenght = Dizendo a quantidade de itens que desejo filtrar no algoritmo
'''
regras = apriori(transacoes, min_support = 0.5, min_confidence = 0.5, min_lenght = 2)
```

```python
regras
```

```python
# Criando a variável para armazenar somente as regras de associação
resultados = list(regras)
resultados
```

```python
# Melhor visualização dos resultados
resultados2 = [list(x) for x in resultados]
resultados2
```
