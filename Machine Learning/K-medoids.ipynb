```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster import cluster_visualizer
```

```python
iris = datasets.load_iris()
```

```python
#Instânciando a classe dos kmedoids
cluster = kmedoids(iris.data[:,0:2], [40,100,149])
#Pegando os centroides
cluster.get_medoids()
```

```python
'''
    Aplicação do algoritmo para agrupamento
'''
cluster.process()
previsoes = cluster.get_clusters()
medoides = cluster.get_medoids()
previsoes
```

```python
v = cluster_visualizer()
v.append_clusters(previsoes, iris.data[:,0:2])
v.append_cluster(medoides, data = iris.data[:,0:2], marker = '*', markersize=20)
v.show()
```

```python
# Criar para gerar duas listas com grupos reais da base de dados e outra com valores dos grupos
# Utilizando posteriormente para visualização da matrix de contingência
lista_prev = []
lista_real = []

for i in range(len(previsoes)):
    for j in range(len(previsoes[i])):
        lista_prev.append(i)
        lista_real.append(iris.target[previsoes[i][j]])
```

```python
#Geração da matrix de contigência, comparando os grupos reais com os grupos previstos
lista_prev = np.asarray(lista_prev)
lista_real = np.asarray(lista_real)
resultados = confusion_matrix(lista_real, lista_prev)
```
