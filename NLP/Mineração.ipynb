```python
import nltk
nltk.download("stopwords") # Baixando as stopwords
import string
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from nltk.corpus import PlaintextCorpusReader
```

```python
#Criação do corpus com base em varios arquivos de texto
corpus =  PlaintextCorpusReader("./drive/MyDrive/Cientista de dados/NLP/Arquivos/Arquivos",  ".*", encoding="ISO-8859-1")
```

```python
arquivos = corpus.fileids()
```

```python
#Acessando texto do primeiro arquivo
texto = corpus.raw("15.txt")
texto
```

```python
#Trazendo todos os texto de cada arquivo
all_text = corpus.raw()
all_text
```

```python
#Obtenção de todas as palavras do corpus e visualização da quantidade
palavras = corpus.words()
print(palavras[170])
len(palavras)
```

```python
#Usando o NLTK obtemos as stepwords em inglês
# stops = stopwords.words("english")
stops = stopwords.words("portuguese")
stops
```

```python
#Criando uma nuvem de palavras
mapa_cores = ListedColormap(["orange","skyblue","crimson", "magenta"])
#Nuvem de palavras com 100 palavras
nuvem = WordCloud(
    background_color="white",
    colormap=mapa_cores,
    max_words=100,
    stopwords=stops
)
#Visualização
nuvem.generate(all_text)
plt.imshow(nuvem)
```

```python
# Criação de nova lista de palavras, removendo stopwords
palavras_semstop = [p for p in palavras if p not in stops]
len(palavras_semstop)
```

```python
#Removendo a pontuação
palavras_sem_pontuacao = [p for p in palavras_semstop if p not in string.punctuation]
len(palavras_sem_pontuacao)
```

```python
# Calculo da frequência de palavras e visualização das mais comuns
frequencia = nltk.FreqDist(palavras_sem_pontuacao)
frequencia
```

```python
mais_comuns = frequencia.most_common()
mais_comuns
```
