```python
import pandas as pd
from sklearn.model_selection import train_test_split # Importando a função de treino e teste para tirar uma amostra do meu DF
from sklearn.preprocessing import LabelEncoder # Codificador de dados categóricos
from sklearn.metrics import confusion_matrix, accuracy_score # Metricas para avaliar o modelo
from sklearn.tree import DecisionTreeClassifier # Importando o classficador de arvore
from sklearn.tree import export_graphviz #Função para visualizar arvores de decisão
```

```python
# Carregando a base de dados e definições das variáveis indepêndentes(X) e depêndente(Y)
credito = pd.read_csv("Credit.csv")
credito.head()
```

```python
previsoes = credito.iloc[:, 0:20].values
classe = credito.iloc[:, -1].values
```

```python
type(previsoes[:,i][0]) == str
```

```python
labelEncoder = LabelEncoder()
for i in range(0,20):
    if type(previsoes[:,i][0]) == str:
        previsoes[:, i] = labelEncoder.fit_transform(previsoes[:,i])
```

```python
previsoes
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsoes, classe, random_state=42, test_size=0.3)
```

```python
#Criação e treinamento de modelo
arvore = DecisionTreeClassifier()
arvore.fit(X_treinamento, y_treinamento)
```

## WebGraphviz
Ferramenta para visualizar o gráfico de arvore de decisão

```python
#Exportação do gráfico para futuras visualizações
export_graphviz(arvore, out_file="tree.dot")
```

```python
#Obtenção das previsões
previsoes = arvore.predict(X_teste)
previsoes
```

```python
confusao = confusion_matrix(y_teste, previsoes)
confusao
```

```python
taxa_de_acerto = accuracy_score(y_teste, previsoes)
taxa_de_acerto
```
