```python
import pandas as pd
from scipy.stats import skewnorm, norm, probplot, shapiro
import matplotlib.pyplot as plt
```

```python
#Usando as funções da boblioteca scipy
#Usando o sub pacote da scipy a biblioteca stats 
#Usando o pacote norm e utilizando a função rvs
dados = norm.rvs(size = 3000)
dados
```

```python
plt.hist(dados)
plt.show()
```

```python
#Verificando se a distribuição é normal
fig, ax = plt.subplots()
probplot(dados, fit=True, plot = ax)
plt.show()
```

```python
# Teste de Shapiro-Wilk
shapiro(dados)
```
