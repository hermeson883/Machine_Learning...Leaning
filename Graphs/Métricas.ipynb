```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_graphml("Grafo.graphml")
```

```python
plt.figure(figsize=(10, 8))
nx.draw(G, pos=nx.spring_layout(G), node_size=100, node_color="skyblue")
plt.show()
```

```python
#Número de relacionamentos de cada nó
nx.degree(G)
```

```python
#Graus de entrada
G.in_degree()
```

```python
G.out_degree()
```

```python
#Grafo com vertice proporcional ao grau
grau = [x[1] for x in G.out_degree]
nx.draw(G, node_size = grau)
```

```python
#Número de vizingo médios
nx.average_neighbor_degree(G, target="in+out")
```

```python
#Vendo se o grafo é isomorfico
G_clone = G.copy()
nx.is_isomorphic(G, G_clone)
```

```python
nx.is_connected(G)
```
