```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
```

```python
#Baixando o modelo
model_name = 'jbochi/madlad400-3b-mt'
model = T5ForConditionalGeneration.from_pretrained(model_name,  device_map="auto")
tokenize = T5Tokenizer.from_pretrained(model_name)
```

```python
# <2pt> === to portugese
# <2de> === to deutschland
# <2es> === to Spanish
# <2en> === to english
# <2fr> === to french
# <2zh> === to chinese
# <2ru> === to Russian
# <2ar> === to Arabic
# <2ja> === to japanese
# <2ko> === to Korean
text = "<2ja> Como esse texto que estou escrevendo ficaria em outra l√≠ngua?"
input_text = tokenize(text, return_tensors='pt').input_ids.to(model.device)
output = model.generate(input_ids = input_text)

tokenize.decode(output[0], skip_special_tokens=True)
```
