```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph([(0,1),(2,3),(0,2),(0,3)])
```

```python
#Atributos do grafo
G.graph["color"] = ["green","red","blue","yellow"]
G.graph["labels"] = {
                        0:"Fernando",
                        1:"Jose",
                        2:"Pedro",
                        3:"Antonio"
                    }
#Plotando o grafo com nomes e cores
nx.draw(G, node_color = G.graph["color"], labels=G.graph["labels"], font_size=20)
```

```python
#Plotando o grafo com as aresta de acordo com o peso
G.graph["weight"] = [1,2,3,4]
nx.draw(G, width=G.graph["weight"], node_color = G.graph["color"], labels=G.graph["labels"])
```

```python
#Grafo com o tamanho do node relativo ao peso da aresta
nx.draw(G, width=G.graph["weight"], node_color = G.graph["color"], labels=G.graph["labels"], node_size=[i * 300 for i in G.graph["weight"]])
```

```python
#Mudando o formato do nodes
nx.draw(G, width=G.graph["weight"], node_color = G.graph["color"], labels=G.graph["labels"], node_size=[i * 300 for i in G.graph["weight"]], node_shape="*")
```
