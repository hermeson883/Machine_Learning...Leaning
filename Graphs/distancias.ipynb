```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph([(0,2),(0,1),(1,2),(2,3),(3,4),(6,7),(3,7),(4,7),(5,6)])
G.graph["labels"] = {0:"A",1:"B",2:"C",3:"D",4:"E",5:"F",6:"G",7:"H"}
G.graph["weight"] = {
    (0,2): 2,
    (0,1): 1,
    (1,2): 2,
    (1,5): 1,
    (2,3): 2,
    (6,7): 1,
    (3,7): 3,
    (4,7): 1,
    (5,6): 1
}
G.graph["color"] = {}
G.graph["color_edges"] = {}

pos = nx.spring_layout(G)

nx.draw(G, pos=pos, labels=G.graph["labels"], with_labels=True, node_size=700, node_color="skyblue", font_size=12, font_weight="bold")

nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=G.graph["weight"])

plt.show()
```

```python
# Pegando o nodes do caminho mais curto
nodes = nx.shortest_path(G, source=0, target = 4)
nodes_names = []
for i in nodes:
    print(G.graph["labels"][i])
    nodes_names.append(G.graph["labels"][i])
```

```python
edges = nx.shortest_path(G, source=0, target = 4)
edges_names = []
path_graph = nx.path_graph(edges)
for ea in path_graph.edges():
    #Pegando o caminho pelo peso das arestas
    # print(ea, G.graph["weight"][ea])
    #Nome do edger do grafo
    edges_names.append(ea)
```

```python
#Colorindo os nodes
for i in G.graph["labels"].values():
    if i in nodes_names:
        print(i)
        G.graph["color"][i] = "green"
    else:
        G.graph["color"][i] = "grey"

# #Colorindo as "edges"
for i in G.graph["weight"].keys():
    print(i)
    if i in edges_names:
        G.graph["color_edges"][i] = "green"
    else:
        G.graph["color_edges"][i] = "gray"
```

```python
nx.draw(G, labels=G.graph["labels"], node_color=G.graph["color"].values(), edge_color = G.graph["color_edges"].values())
```
