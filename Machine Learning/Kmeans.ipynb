```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris = iris

unique, qtd = np.unique(iris.target, return_counts= True)

print(unique, qtd)
```

```python
iris.head()
```

```python
#Agrupamento com K-mean e realizando três clusters
cluster = KMeans(n_clusters=3)
cluster.fit(iris.data)
```

```python
#Centroides
cluster.cluster_centers_
```

```python
previsoes = cluster.labels_
previsoes
```

```python
unicos2, quantidade2 = np.unique(iris.target, return_counts=True)
print(unicos2, quantidade2)
```

```python
resultados = confusion_matrix(iris.target, previsoes)
resultados
```

```python
iris.data[previsoes==0,0]
```

```python
#Gráfico da clusterização
plt.scatter(iris.data[previsoes==0,0], iris.data[previsoes == 0, 1], c = "green", label = "setosa")
plt.scatter(iris.data[previsoes==1,0], iris.data[previsoes == 1, 1], c = "red", label = "Versicolor")
plt.scatter(iris.data[previsoes==2,0], iris.data[previsoes == 2, 1], c = "blue", label = "Virginica")
```
