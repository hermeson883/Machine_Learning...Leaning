```python
import pandas as pd
import numpy as np
from math import ceil
```

```python
#Criação da população da amostra e do valor de K
population = 150
size = 15

#Qual o intervalo que tenho que ter para extração de um novo elemento considerando a base = 150 
# Considerando que quero uma amostra de 15 elementos
k = ceil(population / size)

#Necessito extrair um elemento a cada 10 'saltos'
k
```

```python
r = np.random.randint(low = 1, high = k + 1,size=1)
r
```

```python
acumulador = r[0]
sorteados = []
for i in range(size):
    sorteados.append(acumulador)
    acumulador += k
print(sorteados)
```

```python
len(sorteados)
```

```python
iris = pd.read_csv('iris.csv')
df_final = iris.loc[sorteados]
df_final
```
