```python
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
import pandas as pd
```

```python
iris = datasets.load_iris()
pd.DataFrame(iris.data.T)
```

```python
'''
    data => dados que iremos usar

    C => Número de cluster's

    M => Exponenciação de cada interação

    error => Tolerância de erro

    maxiter => Número máximo de iterações

    init => Começando o algorítmos aleátoriamente
'''
r = skfuzzy.cmeans(data = iris.data.T, c=3, m = 2, error=0.005, maxiter=1000, init = None)
```

```python
#Obtendo as porcentagens de um registro pertencente a um cluster que está na posição 1
previsoes_porcentagem = r[1]
```

```python
previsoes_porcentagem
```

```python
for x in range(150):
    print(previsoes_porcentagem[0][x], previsoes_porcentagem[1][x], previsoes_porcentagem[2][x])
```

```python
previsoes_argmax = previsoes_porcentagem.argmax(axis = 0)
previsoes_argmax
```

```python
previsoes = previsoes_porcentagem.argmax(axis=0)
resultados = confusion_matrix(iris.target, previsoes)
resultados
```
