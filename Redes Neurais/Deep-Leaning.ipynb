```python
dataset = pd.read_csv("Credit2.csv", sep=";")
dataset
```

```python
X = dataset.iloc[:,1:10].values # Separação de base
Y = dataset.iloc[:, 10].values # Pegando nossa coluna categorica
X
```

```python
# Realizando o LabelEncoding da coluna categórica
labelEncoding = LabelEncoder()
X[:,0] = labelEncoding.fit_transform(X[:,0])
X[:,0]
```

```python
# One hot encoding coluna credit_history
#Deve adicionar 5 colunas
"""
    make_column_transformer - Aplicando um transformador no OneHot

    OneHot - Classificando os dados
"""
oneHotEncoder = make_column_transformer((OneHotEncoder(categories="auto", sparse_output=False), [1]), remainder="passthrough")
X =oneHotEncoder.fit_transform(X)
X
```

```python
X.shape
```

```python
#Excluimos a variável para evitar o dummy variable trap
X = X[:, 1:]
X
```

```python
# Label Encoder com classe
labelencoder_Y = LabelEncoder()
y = labelencoder_Y.fit_transform(Y)
y
```

```python
#Treino e teste
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)
print("Tamanho do treino X",len(X_train))
print("Tamanho do teste X",len(X_test))
print("Tamanho do treino Y",len(y_train))
print("Tamanho do teste Y", len(y_test))
```

```python
pd.DataFrame(X)
```

```python
# Padronizando os dados para terém média 0 e desvio padrão 1
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
X_test
```

```python
"""
    Inicializando o modelo

    Dense - Kernel_initializer -> irá começar com pesos uniformes entre (-0.05 e 0.05)
    Dense - input_dim -> Número de informação que irá entrar nos 6 neurônios
    Sequential.compiler ->
    Sequential.fit() -> batch_sizer = Numero de amostras entregues para máquina
"""
classifier = Sequential()
classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu", input_dim = 12))
classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu"))
classifier.add(Dense(units=1, kernel_initializer="uniform", activation="sigmoid"))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ["accuracy"])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
