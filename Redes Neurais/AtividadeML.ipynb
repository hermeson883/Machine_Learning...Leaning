```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("/content/soybean.csv")
df
```

## Removendo as interogações pela moda

```python
# for i in range(36):
#     print(df.iloc[:,i].value_counts())
#     print('#' * 50)

for i in range(36):
    mode = df.iloc[:, i].mode()[0]
    df.iloc[:, i] = df.iloc[:, i].replace("?", mode)

df
```

```python
df["class"].unique()
```

```python
X = df.iloc[:, 0:len(df.columns) - 1].values
X
Y = df.iloc[:, -1].values
Y
```

```python
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
Y
```

```python
_, columns_X = X.shape
columns_X
```

```python
interaction = 0
for i in range(columns_X):
    column_encoder = LabelEncoder()
    X[:, i] = column_encoder.fit_transform(X[:,i])
```

### Verificando a quantidade de colunas está certa

```python
pd.DataFrame(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state=42, test_size = 0.1)
```

```python
model = GaussianNB()

model.fit(X_train, y_train)
```

```python
result = model.predict(X_test)
result
```

```python
accuracy_score(y_test, result)
```

```python
pd.DataFrame(confusion_matrix(y_test, result))
```
